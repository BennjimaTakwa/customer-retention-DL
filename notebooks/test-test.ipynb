```python
import requests
import json

BASE_URL = "https://sentiment-api-8mdq.onrender.com"

# ── 1. Health check
r = requests.get(f"{BASE_URL}/health/")
print("Health:", r.json())

# ── 2. Single prediction
payload = {
    "frustration_score"  : 1.5,
    "engagement_quality" : 0.3,
    "influence_weight"   : 0.6,
    "recency_weight"     : 0.4,
    "word_count"         : 25,
    "has_reply"          : 0,
    "bilstm_prob_neg"    : 0.6,
    "bilstm_prob_neu"    : 0.3,
    "bilstm_prob_pos"    : 0.1,
    "bilstm_confidence"  : 0.75
}

r = requests.post(f"{BASE_URL}/predict/single", json=payload)
print("\nPrediction:")
print(json.dumps(r.json(), indent=2))

# ── 3. Segments list
r = requests.get(f"{BASE_URL}/segments/")
print("\nSegments:")
for seg in r.json():
    print(f"  C{seg['segment_id']} | {seg['segment_name']} | risk={seg['risk_level']}")

# ── 4. Swagger docs
print(f"\n📖 Swagger UI : {BASE_URL}/docs")
print(f"📖 ReDoc      : {BASE_URL}/redoc")
```
